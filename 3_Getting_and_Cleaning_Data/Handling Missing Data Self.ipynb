```python
import pandas as pd
import numpy as np

df = pd.read_csv('melbourne.csv')
df.head()
```

```python
len(df.columns)
```

```python
#There are 21 missing columns, let's see the name of those columns
df.columns
```

```python
df.isnull().sum()
```

```python
#Finding how much percentage of the entire amount of data, 
#these missing values costitute

round(100*(df.isnull().sum())/len(df.index),2)
```

```python
#Removing the columns where the % of missing data is more than 30%

df = df.drop('BuildingArea',axis=1)
df = df.drop('YearBuilt',axis=1)
df = df.drop('CouncilArea',axis=1)
```

```python
round(100*(df.isnull().sum()/len(df.index)),2)
```

## Now we've brought down the % of missing values from something around 50% to 26%

## Now we'll remove the rows containing more than 5 null values

```python
len(df.columns)
```

```python
(df.isnull().sum(axis=1) > 5).sum()
```

```python
# We have 4321 rows which have more than 5 missing values
#Let's find the percentage of them in the total data
sum(df.isnull().sum(axis=1) > 5)/len(df.index) * 100
```

```python
df = df[df.isnull().sum(axis=1) <=5]
```

```python
(df.isnull().sum())/len(df.index) * 100
```

```python
# The price attribute/feature has about 21% of missing values
# Immputing them could lead to bias, so we remove them.
df = df[~np.isnan(df['Price'])]
len(df.index)
```

```python
round(100*(df.isnull().sum()/len(df.index)),2)
```

```python
# Landsize still has 9% of data misisng
# Let's look at whether we could impute them
df['Landsize'].describe()
```

#### Since there is huge difference between the max and the min value and imputing with the mean and median could lead to bias. So we drop these rows

```python
df = df[~np.isnan(df['Landsize'])]
round(100*(df.isnull().sum())/len(df.index),2)
```

```python
df.loc[:,['Lattitude','Longtitude']].describe()
```

```python
#Imputing the values of the Lattitude column with the mean of the column
```

```python
df.loc[np.isnan(df['Lattitude']),['Lattitude']] = df["Lattitude"].mean()
df
```

```python
#Now let's impute the column of Longtitude with its mean
df.loc[np.isnan(df['Longtitude']),['Longtitude']] = df['Longtitude'].mean()
df
```

```python
round(100*(df.isnull().sum()/len(df.index)),2)
```

#### Now we have to fix Car(carparks) and Bathroom, lets begin with carparks

```python
df['Car'].describe()
```

```python
df['Car'].mode()
```

#### Most of the houses have 2 car parkings (which is also known as mode)

```python
# So let's fill the missing values with 2
df.loc[np.isnan(df['Car']),['Car']] = 2
```

#### Now let's get into the Bathroom XO :)

```python
df.Bathroom.describe()
```

```python
#Let's see how many houses have how many bathrooms
df['Bathroom'].value_counts()
```

```python
#Most houses have 1 bathroom, so we'll impute the value with 1
df.loc[np.isnan(df['Bathroom']),['Bathroom']] = 1
df
```

```python
round(100*(df.isnull().sum()/len(df.index)),2)
```
