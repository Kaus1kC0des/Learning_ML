#### If all the stats are taken for both ‘ipl17’ and ‘ipl18’, which team with its total points greater than 25 will have the highest win percentage?

```python
df = pd.merge(ipl18,ipl17,on='Team',suffixes=('_18','_17'))
df['Total Points'] = df['Points_17'] + df['Points_18']
df['Total Matches Won'] = df['Won_17'] + df['Won_18']
df['Total Matches Played'] = df['Matches_17'] + df['Matches_18']
df
```

```python
filterdf = df.sort_values('Total Points',ascending=False)
filterdf
```
