```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,100,11)
y = np.linspace(0,100,11)
plt.plot(x,y,'g*',x,y**2,'b+',x,y**3,'r-')
plt.xlabel('Numbers')
plt.ylabel("Exponents")
plt.title("Plotting Exponents")
plt.show()
```

### Creating Multiple Subplots

```python
plt.figure(1) # 1 is the index or name of the figure
```

```python
x = np.linspace(1,100,10)
y = np.log(x)
```

```python
#subplot 1
plt.subplot(2,2,1) 
#plot contains 2 rows, 2 cols and among those this is the first plot
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel("log(X)")

#subplot 2
plt.subplot(2,2,2)
plt.plot(x,y**2)
plt.xlabel('X')
plt.ylabel("log(X)**2")

#subplot 3
plt.subplot(2,2,3)
plt.plot(x,y**3)
plt.xlabel('X')
plt.ylabel("log(X)**3")

#subplot 4
plt.subplot(2,2,4)
plt.plot(x,y**4)
plt.xlabel('X')
plt.ylabel("log(X)**4")
```

```python
df = pd.read_csv('./global_sales_data/market_fact.csv')
df.head()
```

```python
plt.figure(2)
plt.subplot(121)
plt.boxplot(df['Sales'])

#plotting sales against Log(Sales) as the values appear to be more cluttered
plt.subplot(122)
plt.boxplot(df.Sales)
plt.yscale('log')
plt.show()
```

### Since the above plot is clustered we take the log of the values as the Y scale to see it more clearly in the second plot

### Histograms

```python
plt.hist(df.Sales)
plt.show()
```

```python
plt.hist(df['Sales'])
plt.yscale('log')
plt.show()
```

### In the above example we can clearly see the first histogram does not show the one order which has the value of 89k in Sales column.

### But in the second plot when we plot it against the log of X axis, we can clearly see it

### Scatter Plot shows the distribution/spread of values across the dataset 

```python
plt.scatter(df.Sales,df.Profit)
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title("Sales Vs Profit")


plt.show()
```

# Summary

Box Plots - Displays the spread of the data. Used to visualise the variance in the data, as well as to find information on what proportion of the data is above and below the median value.<br>

Histogram - Displays frequency distribution of a single variable. Used to see what fraction of data is around which values.<br>

Scatter Plot - Displays the scatter of the data. Used to detect outliers. Points to the naturally occurring clusters in the data.
